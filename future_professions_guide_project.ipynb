```python
import pandas as pd
from openagi.planner.task_decomposer import TaskPlanner
from openagi.actions.tools.tavilyqasearch import TavilyWebSearchQA
from openagi.agent import Admin
from openagi.llms.gemini import GeminiModel
```

```python
import os
from getpass import getpass

# setup Gemini and Tavily API Key
os.environ['TAVILY_API_KEY'] = getpass("Enter Tavily API key:")
os.environ['GOOGLE_API_KEY'] = getpass("Enter your Gemini API key:")
os.environ['Gemini_MODEL'] = "gemini-1.5-flash"
os.environ['Gemini_TEMP'] = "0.7"
```

```python
gemini_config = GeminiModel.load_from_env_config()
llm = GeminiModel(config=gemini_config)
# define the planner
plan = TaskPlanner(autonomous=True,human_intervene=True)
```

```python
admin = Admin(
    actions = [TavilyWebSearchQA],
    planner = plan,
    llm = llm,
)
# Enhanced interaction query
res = admin.run(
    query="Please provide a detailed analysis of the top 5 most promising professions in the technology sector over the next 10 years.",
    description=(
        "Focus on the top 5 emerging job roles, required skills, and growth potential within the technology sector for the selected region "
        "over the next decade. Please interactively ask for additional criteria such as region, key growth factors, and data sources before proceeding."
    ),
)
```

```python
print("Response:", res)
```

```python
# Parse the result into a list of dictionaries
professions = []

# Assuming the result is a string, split it into lines and parse accordingly
lines = res.split("\n")
profession = None
for line in lines:
    if line.startswith("1.") or line.startswith("2.") or line.startswith("3.") or line.startswith("4.") or line.startswith("5."):
        if profession is not None:
            professions.append(profession)
        profession = {"Profession": line.split("**")[1], "Description": line.split(":** ")[1]}
    elif profession:
        profession["Description"] += " " + line.strip()

if profession:
    professions.append(profession)

# Convert to a Pandas DataFrame
df = pd.DataFrame(professions)

# Display the DataFrame as a table
print(df)

# Save to CSV
df.to_csv("promising_professions.csv", index=False)

# Save to Excel
df.to_excel("promising_professions.xlsx", index=False)

print("Data saved successfully to 'promising_professions.csv' and 'promising_professions.xlsx'")
```

```python
# Recommendation System
# Collect user input for skills and interests
skills = input("Please enter your main skills (e.g., programming, data analysis, project management): ").lower().strip()
interests = input("Please enter your areas of interest (e.g., artificial intelligence, cybersecurity, cloud computing): ").lower().strip()

# Preprocess user input (remove extra spaces, normalize case)
skills_list = [skill.strip() for skill in skills.split(',')]
interests_list = [interest.strip() for interest in interests.split(',')]

# Filter the professions based on user input
suggested_professions = []
for profession in professions:
    profession_match = 0
    for skill in skills_list:
        if skill in profession["Description"].lower():
            profession_match += 1
    for interest in interests_list:
        if interest in profession["Description"].lower():
            profession_match += 1
    if profession_match > 0:
        suggested_professions.append({"Profession": profession["Profession"], "Description": profession["Description"], "Match Score": profession_match})

# Sort by match score
suggested_professions.sort(key=lambda x: x["Match Score"], reverse=True)

# Display suggested professions
if suggested_professions:
    print("Based on your skills and interests, here are some career suggestions:")
    for suggestion in suggested_professions:
        print(f"- {suggestion['Profession']}: {suggestion['Description']} (Match Score: {suggestion['Match Score']})")
else:
    print("Sorry, no suitable professions were found based on your skills and interests.")
```
